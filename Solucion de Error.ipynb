```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar los datos de prueba
df_e_prueba = pd.read_csv('df_e_prueba.csv', engine='python')
df_r_pru = pd.read_csv('df_r_pru.csv', engine='python')
```

```python
#Cargar los datos completos
df_e1 = pd.read_csv('df_e1.csv', engine='python')
df_r = pd.read_csv('df_r.csv', engine='python')
```

```python
df_r_pru2=df_r_pru.drop(df_r_pru.index[df_r_pru.Id_empleado_nom35 != 21251], axis=0)
suma = df_r_pru2['respuesta'].sum()
suma
```

```python
# Código para iterar a través de las filas del DataFrame de las respuestas e ir sumando las respuestas de cada empleado
# Y asignar la suma total a una nueva columna del DataFrame de los empleados

prueba_e = df_e_prueba

# Agrupar las respuestas por 'Id_empleado_nom35' y calcular la suma
resultados = df_r_pru.groupby('Id_empleado_nom35')['respuesta'].sum().reset_index()

# Renombrar las columnas para que coincidan con el DataFrame de empleados
resultados.columns = ['Id_Empleado', 'Resultado']

# Combinar los resultados con el DataFrame de empleados (prueba_e) utilizando un left join
prueba_e = pd.merge(prueba_e, resultados, on='Id_Empleado', how='left')

# Rellenar cualquier valor NaN con 0
prueba_e['Resultado'].fillna(0, inplace=True)

# Filtrar el DataFrame para mostrar solo 'Id_Empleado' y 'Resultado'
resultados_finales = prueba_e[['Id_Empleado', 'Resultado']]

print(resultados_finales)
```

```python
# Código para iterar a través de las filas del DataFrame de las respuestas e ir sumando las respuestas de cada empleado
# Y asignar la suma total a una nueva columna del DataFrame de los empleados

prueba_e2 = df_e1

# Agrupar las respuestas por 'Id_empleado_nom35' y calcular la suma
resultados = df_r.groupby('Id_empleado_nom35')['respuesta'].sum().reset_index()

# Renombrar las columnas para que coincidan con el DataFrame de empleados
resultados.columns = ['Id_Empleado', 'Resultado']

# Combinar los resultados con el DataFrame de empleados (prueba_e) utilizando un left join
prueba_e2 = pd.merge(prueba_e2, resultados, on='Id_Empleado', how='left')

# Rellenar cualquier valor NaN con 0
prueba_e2['Resultado'].fillna(0, inplace=True)

# Filtrar el DataFrame para mostrar solo 'Id_Empleado' y 'Resultado'
resultados_finales2 = prueba_e2[['Id_Empleado', 'Resultado']]

print(resultados_finales2)
```
